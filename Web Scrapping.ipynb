## Data Cleaning

```python
def data_cleaning(text):
    text = text.lower() # convert everything into lower case
    text = re.sub('[^a-z]',' ',text) #only allowed small a-z
    words = text.split()
    imp_words = [w for w in words if w not in stop]
    return [imp_words]
```

```python
cleaned_data = data_cleaning(master_data)
```

```python
len(cleaned_data[0])
```

```python
word2vec = Word2Vec(cleaned_data,epochs=10,vector_size=300,window=3,min_count=1)
```
